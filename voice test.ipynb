```python
def extract_feature(file_name):
    data, sampling_rate = librosa.load(file_name,sr=16000, duration=1)
    dbpower=librosa.feature.mfcc(y=data, sr=sampling_rate,n_mfcc=20)
    #y, sr = librosa.load(file_name,sr=16000, duration=1)
    #dbpower = librosa.power_to_db(np.abs(librosa.stft(y,n_fft=320))**2, ref=np.max)
    return dbpower

def parse_audio_files(parent_dir,sub_dirs,file_ext='*.wav'):
    featurelist=[]
    dataerror=[]
    features=np.array([])
    featuress, labels = np.empty((0,193)), np.empty(0)
    for label, sub_dir in enumerate(sub_dirs):
        for fn in glob.glob(os.path.join(parent_dir, sub_dir, file_ext)):
            dbpower= extract_feature(fn)
            #print(dbpower.shape)
            if dbpower.shape == (20,32):
                featurelist.append(dbpower)
                labels = np.append(labels,sub_dir)
            else:
                dataerror.append(fn)
    return featurelist, dataerror , np.array(labels, dtype = np.int)

def one_hot_encode(labels):
    n_labels = len(labels)
    n_unique_labels = len(np.unique(labels))
    one_hot_encode = np.zeros((n_labels,n_unique_labels))
    one_hot_encode[np.arange(n_labels), labels] = 1
    return one_hot_encode
```

```python
1 epoch = iterations * batch size

For example, suppose you have 1000 data points in total. You divide these data points into batches of 20. Thus, your batch size is 20. Hence, you have 1000 / 20 = 50 batches in total.

Therefore, in order to process all 1000 data points exactly once (i.e. one epoch) requires 50 iterations (i.e. one iteration per batch).

One epoch is not enough because the model would be underfitted on the data points. Increasing the number of epochs increases the quality of the model until it becomes optimum. However, if you increase the number of epochs any further then it would lead to overfitting of the model on the data points. Hence, you want an optimum number of epochs. There's no rule to determine the optimum number of epochs in advance.﻿ 
```
